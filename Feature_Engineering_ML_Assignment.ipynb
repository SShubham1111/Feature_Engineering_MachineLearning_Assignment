1. What is a parameter?
- A parameter is a variable that defines a model’s behavior or structure.
- In statistics, it represents a characteristic of a population (e.g., mean, variance).
- In machine learning, parameters are learned from data (e.g., weights in linear regression).
- They are adjusted during training to minimize the loss function.

2. What is correlation? What does negative correlation mean?
- Correlation measures the relationship between two variables. It ranges from -1 to +1.
- A positive correlation means both variables increase together.
- A negative correlation means one increases while the other decreases.

3. Define Machine Learning. What are the main components in Machine Learning?
- Machine Learning is a branch of AI that enables systems to learn patterns from data and make predictions.
- Main components:
    - Data – Input for the model.
    - Model – Algorithm to learn patterns.
    - Loss Function – Measures prediction error.
    - Optimizer – Adjusts parameters to minimize loss.

4. How does loss value help in determining whether the model is good or not?
- The loss value measures how far the model’s predictions are from actual values.
- A lower loss indicates better performance.
- If loss decreases during training, the model is learning well. High or stagnant loss may indicate underfitting or poor learning.

5. What are continuous and categorical variables?
- Continuous variables: Numeric values with infinite possible outcomes (e.g., height, weight).
- Categorical variables: Represent discrete categories (e.g., gender, color).
- Continuous data is quantitative categorical is qualitative.
- Continuous can be measured; categorical can be counted.

6. How do we handle categorical variables in Machine Learning? What are the common techniques?
- Categorical variables are converted into numeric form for model use.
- Common techniques:
    - Label Encoding – Assigns each category a numeric label.
    - One-Hot Encoding – Creates binary columns for each category.
    - Ordinal Encoding – Preserves category order if applicable.
    - Target Encoding – Replaces categories with average target values.

7. What do you mean by training and testing a dataset?
- Training set: Used to teach the model and adjust parameters.
- Testing set: Used to evaluate model performance on unseen data.
- Training helps model learn patterns testing checks generalization. Proper split avoids overfitting.

8. What is sklearn.preprocessing?

9. What is a Test set?
- A subset of data not seen by the model during training.
- Used to assess model performance and generalization.
- Helps detect overfitting.
- Usually 20–30% of the dataset.

15. What is causation? Explain difference between correlation and causation with an example.
- Causation means one variable directly affects another.
- Correlation means two variables move together but may not be related.
- Example: Ice cream sales and drowning deaths are correlated but not causally related — both depend on temperature.
- Causation requires proof beyond observed association.

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
- Optimizer updates model parameters to minimize loss.
- Types:
    - SGD (Stochastic Gradient Descent) – Updates weights using gradient of loss.
    - Momentum – Adds moving average to speed up convergence.
    - Adam – Combines momentum + adaptive learning rate; widely used.
    - RMSProp – Adjusts learning rate based on recent gradients.

21. What is feature scaling? How does it help in Machine Learning?
- Feature scaling normalizes data to a common range.
- Prevents large-valued features from dominating small ones.
- Speeds up convergence in optimization algorithms.
- Common methods: Standardization and Min-Max Scaling.

25. Explain data encoding.
- Converts categorical data into numerical format for ML models.
- Prevents models from misinterpreting string labels.
- Common techniques: Label Encoding, One-Hot Encoding, Ordinal Encoding.
- Example: “Red, Blue, Green” → [1, 2, 3] or one-hot vectors.
